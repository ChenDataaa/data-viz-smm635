```python
plt.style.use('your_style')
```

## Bivariate distribution

```python
# simulate data
s = np.random.poisson(10, 1000)
w = 12 + 4 * s + 12 * np.random.randn(1000)
```

```python
plt.scatter(s, w)
```

```python
# initialize a new figure
fig = plt.figure(figsize=(6, 4), facecolor='none')

# add a plot to the figure
ax = plt.subplot(1, 1, 1)

# plot the data
ax.scatter(s, w, marker='*')

# grid
ax.grid(ls='--', which='major', axis='x')

# show the figure 
plt.show()
```
