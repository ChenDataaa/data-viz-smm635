# Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

## Scatter diagram

```python
'''
Let's simulate some data on schooling (years) and wage (£ ,000)
for a sample of 1,000 individuals.

We expect the grand mean of wage (w) is 12; schooling (s) follows
a Poisson distribution with lambda = 10

'''

# let's fix the seed for reproducibility purposes
np.random.seed(123)

# simulation of schooling
s = np.random.poisson(10, 1000)

# simulation of wage - note we're adding a random disturbance
w = 4 * s + 12 * (1 + np.random.randn(1000))
```

```python
# initialize a new figure
fig = plt.figure(figsize=(6, 4), facecolor='none')

# add a plot to the figure
ax = plt.subplot(1, 1, 1)

# plot the data - let's add some transparency
ax.scatter(s, w, marker='o', alpha=0.25)

# we want a grid for the x-axis only
ax.grid(ls='--', which='major', axis='x')

# axes
ax.set_ylabel('Wage (£ ,000)')
ax.set_xlabel('Schooling (years)')

# show the figure 
plt.show()
```

# Surface plots

```python
'''
Surface plots can be used to show the joint distribution of three
variables, X, Y, and Z
'''

# import an additional function for 3d plotting
from mpl_toolkits.mplot3d import axes3d

# import the color map the show the intensity of Z
from matplotlib import cm

# initialize a new figure
fig = plt.figure(figsize=(8, 8))

# add a plot to the figure
ax = fig.gca(projection='3d')

# fake data
X, Y, Z = axes3d.get_test_data(0.05)

# plot the data (change the colormap if you want)
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3)
cset = ax.contour(X, Y, Z, zdir='z', offset=-100, cmap=cm.rainbow)
cset = ax.contour(X, Y, Z, zdir='x', offset=-40, cmap=cm.rainbow)
cset = ax.contour(X, Y, Z, zdir='y', offset=40, cmap=cm.rainbow)

# labels and range of variations for the axes
ax.set_xlabel('X')
ax.set_xlim(-40, 40)
ax.set_ylabel('Y')
ax.set_ylim(-40, 40)
ax.set_zlabel('Z')
ax.set_zlim(-100, 100)

# show the plot
plt.show()
```
